```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import tensorflow as tf
from main import Server
import matplotlib.font_manager as font_manager
from tensorflow.keras.datasets import mnist
spaces = 10
```

```python
server = Server()
server.train_models(epochs=20, alpha=0.2, corrupte=False)
niid_curves = server.get_curve(points=20)
```

```python
server = Server()
server.train_models(epochs=20, alpha=500000, corrupte=False)
iid_curves = server.get_curve(points=20)
```

```python
plt.style.context('bmh')
# font = font_manager.FontProperties(family='Times Roman',
#                                    style='normal', size=12)

font = {'family' : 'DejaVu Sans',
        'size'   : 12}

matplotlib.rc('font', **font)

x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(8, 3))
ax1.plot(iid_curves[0])
ax1.plot(iid_curves[1], 'r')
ax1.set_title('IID')
ax1.set_xticks([0, 0.5, 1]) 
ax1.set_ylabel('Loss')
ax2.plot(niid_curves[0])
ax2.plot(niid_curves[1], 'r')
ax2.set_title('Non-IID')
ax2.set_xticks([0, 0.5, 1]) 
ax3.plot(niid_corrupted_curves[0],label='Train')
ax3.plot(niid_corrupted_curves[1], 'r', label='Test')
ax3.set_title('Corrupted Non-IID')
ax3.set_xticks([0, 0.5, 1]) 
plt.legend(loc='upper right')
ax1.grid(which='major', alpha=0.9, ls = '-.')
ax2.grid(which='major', alpha=0.9, ls = '-.')
ax3.grid(which='major', alpha=0.9, ls = '-.')
plt.savefig('../figs/motivation', bbox_inches='tight')
```
